Check the shape of a WSI

```python
import tifffile as tf
import os

def list_tiff_images_info(img_path): 
    for entry in os.listdir(img_path):
        full_path = os.path.join(img_path, entry)
        if os.path.isfile(full_path) and entry.lower().endswith('.tiff'):
            img = tf.imread(full_path)
            print(img.shape)

img_path = "minisample/train_images"
list_tiff_images_info(img_path)
```

Check a group of patches

```python
from PIL import Image

def convert_tiff_to_jpeg(tiff_path, jpeg_path):
    Image.MAX_IMAGE_PIXELS = None
    with Image.open(tiff_path) as img:
        img.save(jpeg_path, "JPEG")

convert_tiff_to_jpeg("minisample/train_images/0a5d6a20d1429dd55b94ae1b857ad573.tiff", "example.jpeg")
```

```python
from PIL import Image, ImageEnhance

def darken_unextracted_areas(original_img_path, tiles_dir, num_rows, num_cols, tile_width, tile_height):
    original_img = Image.open(original_img_path)

    enhancer = ImageEnhance.Brightness(original_img)
    darkened_img = enhancer.enhance(0.7)

    for row in range(num_rows):
        for col in range(num_cols):
            tile_path = f"{tiles_dir}/{col + 1}_{row + 1}.jpeg"
            try:
                tile_img = Image.open(tile_path)
                left = col * tile_width
                upper = row * tile_height
                
                darkened_img.paste(tile_img, (left, upper))
                print(f"find {tile_path}")
            except FileNotFoundError:
                continue

    darkened_img.save("example_new.jpg")

darken_unextracted_areas("example.jpeg", "patch/0a5d6a20d1429dd55b94ae1b857ad573_files/1.0", num_rows=19, num_cols=30, tile_width=512, tile_height=512)
```
